```python
import pandas as pd

df_2024 = pd.read_csv('/content/2024plays.csv')
df_2023 = pd.read_csv('/content/2023plays.csv')
df_2022 = pd.read_csv('/content/2022plays.csv')

display(df_2022.head())
```

```python
import pandas as pd

leverage_df = pd.read_csv('/content/leverage.txt', names=['Team_At_Bat', 'Inning', 'Outs', 'Baserunners', 'Score_Margin', 'Leverage_Index'])
display(leverage_df.head())
```

```python
def get_baserunner_representation(row):
    br1 = row['br1_pre']
    br2 = row['br2_pre']
    br3 = row['br3_pre']

    if pd.isna(br1) and pd.isna(br2) and pd.isna(br3):
        return 1
    elif pd.notna(br1) and pd.isna(br2) and pd.isna(br3):
        return 2
    elif pd.isna(br1) and pd.notna(br2) and pd.isna(br3):
        return 3
    elif pd.isna(br1) and pd.isna(br2) and pd.notna(br3):
        return 4
    elif pd.notna(br1) and pd.notna(br2) and pd.isna(br3):
        return 5
    elif pd.notna(br1) and pd.isna(br2) and pd.notna(br3):
        return 6
    elif pd.isna(br1) and pd.notna(br2) and pd.notna(br3):
        return 7
    elif pd.notna(br1) and pd.notna(br2) and pd.notna(br3):
        return 8
    else:
        return 0 # Should not happen based on the logic

df_2024['Baserunners'] = df_2024.apply(get_baserunner_representation, axis=1)
df_2023['Baserunners'] = df_2023.apply(get_baserunner_representation, axis=1)
df_2022['Baserunners'] = df_2022.apply(get_baserunner_representation, axis=1)



display(df_2024[['br1_pre', 'br2_pre', 'br3_pre', 'Baserunners']].head())
```

```python
df_2024['visiting_score'] = df_2024.apply(lambda row: row['runs'] if row['vis_home'] == 0 else 0, axis=1)
df_2024['home_score'] = df_2024.apply(lambda row: row['runs'] if row['vis_home'] == 1 else 0, axis=1)

df_2024['cumulative_visiting_score'] = df_2024.groupby('gid')['visiting_score'].cumsum()
df_2024['cumulative_home_score'] = df_2024.groupby('gid')['home_score'].cumsum()

df_2024['Score_Margin'] = df_2024['cumulative_visiting_score'] - df_2024['cumulative_home_score']

df_2023['visiting_score'] = df_2023.apply(lambda row: row['runs'] if row['vis_home'] == 0 else 0, axis=1)
df_2023['home_score'] = df_2023.apply(lambda row: row['runs'] if row['vis_home'] == 1 else 0, axis=1)

df_2023['cumulative_visiting_score'] = df_2023.groupby('gid')['visiting_score'].cumsum()
df_2023['cumulative_home_score'] = df_2023.groupby('gid')['home_score'].cumsum()

df_2023['Score_Margin'] = df_2023['cumulative_visiting_score'] - df_2023['cumulative_home_score']

df_2022['visiting_score'] = df_2022.apply(lambda row: row['runs'] if row['vis_home'] == 0 else 0, axis=1)
df_2022['home_score'] = df_2022.apply(lambda row: row['runs'] if row['vis_home'] == 1 else 0, axis=1)

df_2022['cumulative_visiting_score'] = df_2022.groupby('gid')['visiting_score'].cumsum()
df_2022['cumulative_home_score'] = df_2022.groupby('gid')['home_score'].cumsum()

df_2022['Score_Margin'] = df_2022['cumulative_visiting_score'] - df_2022['cumulative_home_score']

display(df_2023[['gid', 'inning', 'top_bot', 'vis_home', 'runs', 'visiting_score', 'home_score', 'cumulative_visiting_score', 'cumulative_home_score', 'Score_Margin']].head())
```

```python
# Rename columns in each dataframe to include the year
df_2024 = df_2024.rename(columns={
    'inning': 'inning_2024',
    'outs_pre': 'outs_pre_2024',
    'Baserunners': 'Baserunners_2024',
    'Score_Margin': 'Score_Margin_2024',
    'Team_At_Bat': 'Team_At_Bat_2024',
    'slugging_contribution_per_ab': 'slugging_contribution_per_ab_2024',
    'obp_contribution_per_play': 'obp_contribution_per_play_2024'
})

df_2023 = df_2023.rename(columns={
    'inning': 'inning_2023',
    'outs_pre': 'outs_pre_2023',
    'Baserunners': 'Baserunners_2023',
    'Score_Margin': 'Score_Margin_2023',
    'Team_At_Bat': 'Team_At_Bat_2023',
    'slugging_contribution_per_ab': 'slugging_contribution_per_ab_2023',
    'obp_contribution_per_play': 'obp_contribution_per_play_2023'
})

df_2022 = df_2022.rename(columns={
    'inning': 'inning_2022',
    'outs_pre': 'outs_pre_2022',
    'Baserunners': 'Baserunners_2022',
    'Score_Margin': 'Score_Margin_2022',
    'Team_At_Bat': 'Team_At_Bat_2022',
    'slugging_contribution_per_ab': 'slugging_contribution_per_ab_2022',
    'obp_contribution_per_play': 'obp_contribution_per_play_2022'
})

# Merge with leverage_df for each year, using renamed columns and corresponding leverage columns
merged_2024 = pd.merge(
    df_2024,
    leverage_df,
    left_on=['inning_2024', 'outs_pre_2024', 'Baserunners_2024', 'Score_Margin_2024', 'Team_At_Bat_2024'],
    right_on=['Inning', 'Outs', 'Baserunners', 'Score_Margin', 'Team_At_Bat'],
    how='inner'
)

merged_2023 = pd.merge(
    df_2023,
    leverage_df,
    left_on=['inning_2023', 'outs_pre_2023', 'Baserunners_2023', 'Score_Margin_2023', 'Team_At_Bat_2023'],
    right_on=['Inning', 'Outs', 'Baserunners', 'Score_Margin', 'Team_At_Bat'],
    how='inner'
)

merged_2022 = pd.merge(
    df_2022,
    leverage_df,
    left_on=['inning_2022', 'outs_pre_2022', 'Baserunners_2022', 'Score_Margin_2022', 'Team_At_Bat_2022'],
    right_on=['Inning', 'Outs', 'Baserunners', 'Score_Margin', 'Team_At_Bat'],
    how='inner'
)

# Display the head of one of the merged dataframes to verify
display(merged_2024.head())
```

```python
# Calculate total bases for each play
df_2024['total_bases'] = df_2024['single'] * 1 + df_2024['double'] * 2 + df_2024['triple'] * 3 + df_2024['hr'] * 4

# Calculate slugging contribution per at bat, handling division by zero
# Use .loc to avoid SettingWithCopyWarning
df_2024.loc[df_2024['ab'] > 0, 'slugging_contribution_per_ab'] = df_2024['total_bases'] / df_2024['ab']
df_2024.loc[df_2024['ab'] == 0, 'slugging_contribution_per_ab'] = pd.NA

#Repeat for 2023 and 2022
df_2023['total_bases'] = df_2023['single'] * 1 + df_2023['double'] * 2 + df_2023['triple'] * 3 + df_2023['hr'] * 4
df_2023.loc[df_2023['ab'] > 0, 'slugging_contribution_per_ab'] = df_2023['total_bases'] / df_2023['ab']
df_2023.loc[df_2023['ab'] == 0, 'slugging_contribution_per_ab'] = pd.NA

df_2022['total_bases'] = df_2022['single'] * 1 + df_2022['double'] * 2 + df_2022['triple'] * 3 + df_2022['hr'] * 4
df_2022.loc[df_2022['ab'] > 0, 'slugging_contribution_per_ab'] = df_2022['total_bases'] / df_2022['ab']
df_2022.loc[df_2022['ab'] == 0, 'slugging_contribution_per_ab'] = pd.NA

# Display the relevant columns to verify the calculation
display(df_2024[['single', 'double', 'triple', 'hr', 'ab', 'total_bases', 'slugging_contribution_per_ab']].head())
```

```python
# Calculate on-base components: hits, walks, and hit-by-pitch
df_2024['obp_components'] = df_2024['single'] + df_2024['double'] + df_2024['triple'] + df_2024['hr'] + df_2024['walk'] + df_2024['hbp']
# Calculate the OBP denominator: at-bats, walks, hit-by-pitch, and sacrifice flies
df_2024['obp_denominator'] = df_2024['ab'] + df_2024['walk'] + df_2024['hbp'] + df_2024['sf']
# Calculate on-base percentage contribution per play, handling division by zero
# Use .loc to avoid SettingWithCopyWarning
df_2024.loc[df_2024['obp_denominator'] > 0, 'obp_contribution_per_play'] = df_2024['obp_components'] / df_2024['obp_denominator']
df_2024.loc[df_2024['obp_denominator'] == 0, 'obp_contribution_per_play'] = pd.NA

#repeat for 2023 and 2022

df_2023['obp_components'] = df_2023['single'] + df_2023['double'] + df_2023['triple'] + df_2023['hr'] + df_2023['walk'] + df_2023['hbp']
df_2023['obp_denominator'] = df_2023['ab'] + df_2023['walk'] + df_2023['hbp'] + df_2023['sf']
df_2023.loc[df_2023['obp_denominator'] > 0, 'obp_contribution_per_play'] = df_2023['obp_components'] / df_2023['obp_denominator']
df_2023.loc[df_2023['obp_denominator'] == 0, 'obp_contribution_per_play'] = pd.NA

df_2022['obp_components'] = df_2022['single'] + df_2022['double'] + df_2022['triple'] + df_2022['hr'] + df_2022['walk'] + df_2022['hbp']
df_2022['obp_denominator'] = df_2022['ab'] + df_2022['walk'] + df_2022['hbp'] + df_2022['sf']
df_2022.loc[df_2022['obp_denominator'] > 0, 'obp_contribution_per_play'] = df_2022['obp_components'] / df_2022['obp_denominator']
df_2022.loc[df_2022['obp_denominator'] == 0, 'obp_contribution_per_play'] = pd.NA

# Display the relevant columns to verify the calculation
display(df_2024[['single', 'double', 'triple', 'hr', 'walk', 'hbp', 'sf', 'ab', 'obp_components', 'obp_denominator', 'obp_contribution_per_play']].head())
```

```python
# Calculate levered slugging and levered obp by multiplying contribution by leverage index
merged_2024['levered_slug_per_ab'] = merged_2024['slugging_contribution_per_ab_2024'] * merged_2024['Leverage_Index']
merged_2024['levered_obp'] = merged_2024['obp_contribution_per_play_2024'] * merged_2024['Leverage_Index']

merged_2023['levered_slug_per_ab'] = merged_2023['slugging_contribution_per_ab_2023'] * merged_2023['Leverage_Index']
merged_2023['levered_obp'] = merged_2023['obp_contribution_per_play_2023'] * merged_2023['Leverage_Index']

merged_2022['levered_slug_per_ab'] = merged_2022['slugging_contribution_per_ab_2022'] * merged_2022['Leverage_Index']
merged_2022['levered_obp'] = merged_2022['obp_contribution_per_play_2022'] * merged_2022['Leverage_Index']

# Display the relevant columns to verify the calculation for one of the merged dataframes
display(merged_2024[['slugging_contribution_per_ab_2024', 'obp_contribution_per_play_2024', 'Leverage_Index', 'levered_slug_per_ab', 'levered_obp']].head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

##2024

# Merge average traditional OPS, average levered OBP, and average levered slugging for 2024
players_avg_ops_2024 = pd.merge(player_ops_avg_2024[['batter', 'avg_ops_2024']], player_levered_obp_avg_2024[['batter', 'avg_levered_obp_2024']], on='batter', how='left')
players_avg_ops_2024 = pd.merge(players_avg_ops_2024, player_levered_slug_avg_2024[['batter', 'avg_levered_slug_2024']], on='batter', how='left')

# Calculate average levered OPS for 2024
players_avg_ops_2024['avg_levered_ops_2024'] = players_avg_ops_2024['avg_levered_obp_2024'] + players_avg_ops_2024['avg_levered_slug_2024']

# Merge with player_total_ab to get total_ab for 2024
player_year_ab_2024 = player_total_ab[['batter', 'ab_2024']].rename(columns={'ab_2024': 'total_ab'}) # Use player_total_ab
players_avg_ops_filtered_2024 = pd.merge(players_avg_ops_2024, player_year_ab_2024, on='batter', how='left')


# Filter for players with more than 300 at-bats in 2024
cutoff = 300
players_avg_ops_filtered_2024 = players_avg_ops_filtered_2024[players_avg_ops_filtered_2024['total_ab'] > cutoff].copy()
n = len(players_avg_ops_filtered_2024)


# Remove rows with NaN values in either 'avg_ops_2024' or 'avg_levered_ops_2024' for regression
players_avg_ops_filtered_for_regression_2024 = players_avg_ops_filtered_2024.dropna(subset=['avg_ops_2024', 'avg_levered_ops_2024']).copy()

# Calculate the line of best fit for 2024
slope_2024, intercept_2024, r_value_2024, p_value_2024, std_err_2024 = linregress(players_avg_ops_filtered_for_regression_2024['avg_ops_2024'], players_avg_ops_filtered_for_regression_2024['avg_levered_ops_2024'])
line_2024 = slope_2024 * players_avg_ops_filtered_for_regression_2024['avg_ops_2024'] + intercept_2024

plt.figure(figsize=(10, 6))
plt.scatter(players_avg_ops_filtered_2024['avg_ops_2024'], players_avg_ops_filtered_2024['avg_levered_ops_2024'], alpha=0.5)
plt.plot(players_avg_ops_filtered_for_regression_2024['avg_ops_2024'], line_2024, color='red', label=f'Line of Best Fit 2024 (R={r_value_2024:.2f})') # Add the line of best fit
plt.xlabel('Traditional OPS (2024)')
plt.ylabel('Levered OPS (2024)')
plt.title(f'Traditional OPS vs. Levered OPS by Player (2024, > {cutoff} ABs, n = {n})')
plt.grid(True)
plt.legend() # Add a legend to show the label for the line
plt.show()

# Calculate predicted levered OPS based on the line of best fit for 2024
players_avg_ops_filtered_2024['predicted_levered_ops_2024'] = slope_2024 * players_avg_ops_filtered_2024['avg_ops_2024'] + intercept_2024

# Calculate the residuals (actual levered OPS - predicted levered OPS) for 2024
players_avg_ops_filtered_2024['residual_2024'] = players_avg_ops_filtered_2024['avg_levered_ops_2024'] - players_avg_ops_filtered_2024['predicted_levered_ops_2024']

# Store the residuals with the player's name for 2024
player_residuals_2024 = players_avg_ops_filtered_2024[['batter', 'residual_2024']].copy()

# Display the residuals for 2024
display(player_residuals_2024.head())
```
